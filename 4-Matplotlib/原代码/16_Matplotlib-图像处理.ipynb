### Matplotlib-图像处理

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 读取图片
- imread

```python
img = plt.imread('data/pandas.jpeg')
img
```

### 显示图片
- imshow

```python
plt.imshow(img)
plt.savefig('images/7-1.png')
```

```python
# 垂直翻转显示
plt.imshow(img, origin='lower')

# 或
plt.imshow(img[::-1])

plt.savefig('images/7-2.png')
```

```python
# 水平翻转
plt.imshow(img[:, ::-1])

plt.savefig('images/7-3.png')
```

### 保存图片
- imsave

```python
# 保存左右翻转后的图片
plt.imsave('pandas.png', img[:, ::-1])
```
